# Ejercicio 1

```python
# Generar tablas de multiplicar del 1 al 5
for numero in range(1, 6):
    print(f"Tabla de multiplicar del {numero}:")

    # Bucle para los multiplicadores del 1 al 10
    for multiplicador in range(1, 11):
        producto = numero * multiplicador

        # Utilizar 'continue' para omitir multiplicadores pares
        if multiplicador % 2 == 0:
            continue  # Salta al siguiente ciclo del bucle interno

        # Utilizar 'break' para detener la tabla si el producto supera 20
        if producto > 20:
            print("  ... Producto mayor que 20, deteniendo la tabla.")
            break  # Sale del bucle interno y continúa con el siguiente número

        # Utilizar 'pass' para ilustrar su uso sin afectar la lógica
        if multiplicador == 5:
            pass  # No hace nada, solo sirve como marcador de posición

        # Imprimir el resultado de la multiplicación
        print(f"{numero} x {multiplicador} = {producto}")

    # Línea separadora para mayor claridad
    print("-" * 30)
```

# Ejercicio 2

```python
def process_text(text="This is a default text with Python and amazing words.", replace_words=["Python", "amazing"]):
    # Convertir el texto a minúsculas
    text = text.lower()

    # Eliminar espacios en blanco iniciales y finales
    text = text.strip()

    # Contador de palabras reemplazadas
    replaced_count = 0

    # Reemplazar las palabras de la lista por asteriscos
    for word in replace_words:
        # Reemplazar las coincidencias y contar cuántas veces fue reemplazada
        if word.lower() in text:
            text = text.replace(word.lower(), '*' * len(word))
            replaced_count += 1

    # Retornar el texto procesado y la cantidad de palabras reemplazadas
    return text, replaced_count

def main():
    # Llamar a la función process_text con los valores predeterminados
    processed_text, replaced_count = process_text()

    # Imprimir los resultados
    print(f'Processed text: "{processed_text}"')
    print(f'Palabras reemplazadas: {replaced_count}')

# Ejecutar la función principal
if __name__ == "__main__":
    main()
```

# Ejercicio 3



```python
def is_prime(n):
    """
    Comprueba si un número n es primo.
    Un número primo es aquel que solo tiene dos divisores: 1 y él mismo.

    Args:
    n (int): El número a verificar.

    Returns:
    bool: True si n es primo, False en caso contrario.
    """
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

def prime_list(limit):
    """
    Genera una lista de números primos hasta un límite determinado.

    Args:
    limit (int): El límite superior para la generación de números primos.

    Returns:
    list: Lista de números primos hasta el límite.
    """
    primes = []
    for num in range(2, limit + 1):
        if is_prime(num):
            primes.append(num)
    return primes

def check_palindrome(primes):
    """
    Busca palíndromos en una lista de números primos.
    Un número palíndromo se lee igual de adelante hacia atrás.

    Args:
    primes (list): Lista de números primos.

    Returns:
    list: Lista de números primos que son palíndromos.
    """
    palindromes = []
    for prime in primes:
        if str(prime) == str(prime)[::-1]:
            palindromes.append(prime)
    return palindromes

def categorize_prime(prime):
    """
    Clasifica los números primos en tres categorías:
    - "Pequeños" para números menores que 10.
    - "Medianos" para números entre 10 y 99.
    - "Grandes" para números de 100 o más.

    Args:
    prime (int): El número primo a clasificar.

    Returns:
    str: Categoría del número primo ("pequeños", "medianos", "grandes").
    """
    if prime < 10:
        return "pequeños"
    elif 10 <= prime <= 99:
        return "medianos"
    else:
        return "grandes"

def main():
    """
    Función principal para probar las funciones de análisis de números primos.
    """
    # Definir un límite para la generación de números primos
    limit = 150
    print(f"Generando lista de números primos hasta {limit}...")
    primes = prime_list(limit)

    # Mostrar la lista de primos generada
    print(f"Números primos hasta {limit}: {primes}")

    # Buscar y mostrar palíndromos en la lista de números primos
    palindromes = check_palindrome(primes)
    print(f"Primos que son palíndromos: {palindromes}")

    # Clasificar cada número primo
    print("Clasificación de números primos:")
    for prime in primes:
        category = categorize_prime(prime)
        print(f"El primo {prime} es {category}.")

# Ejecutar la función principal
if __name__ == "__main__":
    main()
```

# Ejercicio 4

```python
def suma_total(numeros):
    """
    Calcula la suma de todos los números en la lista.

    Args:
    numeros (list): Lista de números enteros.

    Returns:
    int: La suma de todos los números.
    """
    return sum(numeros)

def suma_positivos_negativos(numeros):
    """
    Calcula la suma de los números positivos y negativos por separado.

    Args:
    numeros (list): Lista de números enteros.

    Returns:
    tuple: Suma de los positivos y suma de los negativos.
    """
    suma_positivos = sum(num for num in numeros if num > 0)
    suma_negativos = sum(num for num in numeros if num < 0)
    return suma_positivos, suma_negativos

def max_min(numeros):
    """
    Encuentra el número máximo y el mínimo de la lista.

    Args:
    numeros (list): Lista de números enteros.

    Returns:
    tuple: El número máximo y el mínimo.
    """
    return max(numeros), min(numeros)

def son_unicos(numeros):
    """
    Comprueba si todos los números de la lista son únicos.

    Args:
    numeros (list): Lista de números enteros.

    Returns:
    bool: True si todos los números son únicos, False en caso contrario.
    """
    return len(numeros) == len(set(numeros))

def positivos_vs_negativos(numeros):
    """
    Comprueba si hay más números positivos o negativos en la lista y devuelve un mensaje.

    Args:
    numeros (list): Lista de números enteros.

    Returns:
    str: Mensaje indicando si hay más números positivos, negativos o si están equilibrados.
    """
    positivos = sum(1 for num in numeros if num > 0)
    negativos = sum(1 for num in numeros if num < 0)

    if positivos > negativos:
        return "La lista contiene más números positivos."
    elif negativos > positivos:
        return "La lista contiene más números negativos."
    else:
        return "La lista tiene la misma cantidad de números positivos y negativos."

def main():
    """
    Función principal que gestiona el menú y las operaciones sobre la lista de números.
    """
    # Leer la lista de números enteros
    numeros = list(map(int, input("Introduce una lista de números enteros separados por espacios: ").split()))

    while True:
        # Menú de opciones
        print("\nMenú de operaciones:")
        print("1. Calcular la suma de todos los números.")
        print("2. Calcular la suma de los números positivos y la suma de los negativos.")
        print("3. Encontrar el número máximo y mínimo de la lista.")
        print("4. Comprobar si todos los números de la lista son únicos.")
        print("5. Mostrar si hay más números positivos o negativos.")
        print("6. Salir.")

        # Pedir opción al usuario
        opcion = input("Elige una opción (1-6): ")

        if opcion == "1":
            # Opción 1: Suma de todos los números
            total = suma_total(numeros)
            print(f"La suma total de los números es: {total}")

        elif opcion == "2":
            # Opción 2: Suma de positivos y negativos
            suma_positivos, suma_negativos = suma_positivos_negativos(numeros)
            print(f"Suma de números positivos: {suma_positivos}")
            print(f"Suma de números negativos: {suma_negativos}")

        elif opcion == "3":
            # Opción 3: Número máximo y mínimo
            maximo, minimo = max_min(numeros)
            print(f"El número máximo es: {maximo}")
            print(f"El número mínimo es: {minimo}")

        elif opcion == "4":
            # Opción 4: Comprobar si los números son únicos
            if son_unicos(numeros):
                print("Todos los números de la lista son únicos.")
            else:
                print("Hay números repetidos en la lista.")

        elif opcion == "5":
            # Opción 5: Más números positivos o negativos
            mensaje = positivos_vs_negativos(numeros)
            print(mensaje)

        elif opcion == "6":
            # Opción 6: Salir
            print("Saliendo del programa.")
            break

        else:
            print("Opción no válida, intenta de nuevo.")

# Ejecutar la función principal
if __name__ == "__main__":
    main()
```

# Ejercicio 5

```python
# Variable global que almacena la lista de tareas
tareas = []

def add_task(task):
    """
    Agrega una nueva tarea a la lista global de tareas.

    Args:
        task (str): La tarea a agregar.
    """
    global tareas
    tareas.append(task)
    print(f"Tarea '{task}' agregada.")

def remove_task(task):
    """
    Elimina una tarea de la lista global de tareas si existe.

    Args:
        task (str): La tarea a eliminar.
    """
    global tareas
    try:
        tareas.remove(task)
        print(f"Tarea '{task}' eliminada.")
    except ValueError:
        print(f"Tarea '{task}' no se encuentra en la lista.")

def list_tasks():
    """
    Enumera todas las tareas almacenadas en la lista global de tareas.
    """
    global tareas
    if not tareas:
        print("La lista de tareas está vacía.")
    else:
        print("Lista de tareas:")
        for idx, tarea in enumerate(tareas, 1):
            print(f"{idx}. {tarea}")

def task_manager():
    """
    Administra las operaciones de agregar, modificar y enumerar tareas.
    Incluye una función interna para modificar la primera tarea de la lista.
    """
    # Variable local para contar modificaciones
    contador_modificaciones = 0

    def modify_task():
        """
        Modifica la primera tarea en la lista de tareas.
        Utiliza la palabra clave 'nonlocal' para modificar la variable local 'contador_modificaciones'.
        """
        nonlocal contador_modificaciones
        global tareas
        if tareas:
            original = tareas[0]
            modificado = f"{original} (modificada)"
            tareas[0] = modificado
            contador_modificaciones += 1
            print(f"Tarea '{original}' modificada a '{modificado}'.")
            print(f"Total de modificaciones: {contador_modificaciones}")
        else:
            print("No hay tareas para modificar.")

    # Llamada a la función interna para modificar la tarea
    modify_task()

def main():
    """
    Función principal que demuestra el uso de las funciones de gestión de tareas.
    """
    # Agregar tareas
    add_task("Comprar leche")
    add_task("Llamar al doctor")
    add_task("Hacer ejercicio")

    # Enumerar tareas
    list_tasks()

    # Modificar la primera tarea
    task_manager()

    # Enumerar tareas después de la modificación
    list_tasks()

    # Eliminar una tarea
    remove_task("Hacer ejercicio")

    # Enumerar tareas después de la eliminación
    list_tasks()

if __name__ == "__main__":
    main()

```

# Ejercicio 6

```python
def imprimir_cadenas_con_letra(cadenas, letra):
    """
    Imprime las cadenas de la lista que comienzan con una letra específica.

    Args:
        cadenas (list): Lista de cadenas.
        letra (str): Letra con la que deben comenzar las cadenas.
    """
    print(f"Cadenas que comienzan con la letra '{letra}':")
    for cadena in cadenas:
        if cadena.startswith(letra):
            print(cadena)

def contar_cadenas_con_subcadena(cadenas, subcadena):
    """
    Cuenta cuántas cadenas contienen una subcadena específica.

    Args:
        cadenas (list): Lista de cadenas.
        subcadena (str): Subcadena a buscar en las cadenas.

    Returns:
        int: Número de cadenas que contienen la subcadena.
    """
    contador = sum(1 for cadena in cadenas if subcadena in cadena)
    return contador

def encontrar_cadena_mas_larga_y_corta(cadenas):
    """
    Encuentra la cadena más larga y la más corta en la lista.

    Args:
        cadenas (list): Lista de cadenas.

    Returns:
        tuple: La cadena más larga y la más corta.
    """
    cadena_mas_larga = max(cadenas, key=len)
    cadena_mas_corta = min(cadenas, key=len)
    return cadena_mas_larga, cadena_mas_corta

def verificar_si_son_mismo_objeto(cadena1, cadena2):
    """
    Verifica si dos cadenas son el mismo objeto en memoria utilizando el operador 'is'.

    Args:
        cadena1 (str): Primera cadena.
        cadena2 (str): Segunda cadena.

    Returns:
        bool: True si son el mismo objeto, False en caso contrario.
    """
    return cadena1 is cadena2

def verificar_cadenas_mayores_de_10(cadenas):
    """
    Utiliza un bucle for...else para verificar si alguna cadena tiene más de 10 caracteres.

    Args:
        cadenas (list): Lista de cadenas.
    """
    for cadena in cadenas:
        if len(cadena) > 10:
            print(f"La cadena '{cadena}' tiene más de 10 caracteres.")
            break
    else:
        print("Ninguna cadena tiene más de 10 caracteres.")

def main():
    """
    Función principal que gestiona el menú y las operaciones sobre la lista de cadenas.
    """
    # Leer la lista de cadenas desde teclado
    cadenas = input("Introduce una lista de cadenas separadas por comas: ").split(',')

    while True:
        # Menú de opciones
        print("\nMenú de operaciones:")
        print("1. Imprimir cadenas que comiencen con una letra específica.")
        print("2. Contar cuántas cadenas contienen una subcadena específica.")
        print("3. Encontrar la cadena más larga y la más corta.")
        print("4. Verificar si dos cadenas son el mismo objeto.")
        print("5. Verificar si alguna cadena tiene más de 10 caracteres.")
        print("6. Salir.")

        # Pedir opción al usuario
        opcion = input("Elige una opción (1-6): ")

        if opcion == "1":
            # Opción 1: Imprimir cadenas que comiencen con una letra específica
            letra = input("Introduce la letra: ")
            imprimir_cadenas_con_letra(cadenas, letra)

        elif opcion == "2":
            # Opción 2: Contar cuántas cadenas contienen una subcadena específica
            subcadena = input("Introduce la subcadena a buscar: ")
            conteo = contar_cadenas_con_subcadena(cadenas, subcadena)
            print(f"Hay {conteo} cadenas que contienen la subcadena '{subcadena}'.")

        elif opcion == "3":
            # Opción 3: Encontrar la cadena más larga y la más corta
            cadena_mas_larga, cadena_mas_corta = encontrar_cadena_mas_larga_y_corta(cadenas)
            print(f"La cadena más larga es: '{cadena_mas_larga}'")
            print(f"La cadena más corta es: '{cadena_mas_corta}'")

        elif opcion == "4":
            # Opción 4: Verificar si dos cadenas son el mismo objeto
            cadena1 = input("Introduce la primera cadena: ")
            cadena2 = input("Introduce la segunda cadena: ")
            if verificar_si_son_mismo_objeto(cadena1, cadena2):
                print("Las cadenas son el mismo objeto en memoria.")
            else:
                print("Las cadenas NO son el mismo objeto en memoria.")

        elif opcion == "5":
            # Opción 5: Verificar si alguna cadena tiene más de 10 caracteres
            verificar_cadenas_mayores_de_10(cadenas)

        elif opcion == "6":
            # Opción 6: Salir
            print("Saliendo del programa.")
            break

        else:
            print("Opción no válida, intenta de nuevo.")

# Ejecutar la función principal
if __name__ == "__main__":
    main()
```
